```python
import cromo
from cromo.click_encapsulate.commands import find_models
from cromo.catalogs.model_catalog import checkConfigViability, getModelRules
```

```python
RULES_DIR='/Users/varun/git/cromo/src/cromo/resources/rules' 
ONTOLOGY_DIR='/Users/varun/git/cromo/src/cromo/resources/ontology'
```

```python
scenario = 'Fire'
region_geojson = '/Users/varun/git/cromo/src/cromo/test/georgetown.geojson'
start_date = '2020-10-01T00:00:00'
end_date = '2020-10-31T00:00:00'
```

```python
import geojsonio
import geopandas as gpd
region = gpd.read_file(region_geojson)
geojsonio.display(region.to_json())
```

```python
configs = find_models(scenario)
for config in configs:
        if config.has_input is not None:
            rules = getModelRules(config.id, rulesdir=RULES_DIR)
            if len(rules) > 0:
                print("\n{}\n{}".format(config.label[0], "="*len(config.label[0])))
                checkConfigViability(config, region_geojson, start_date, end_date, rulesdir=RULES_DIR, ontdir=ONTOLOGY_DIR)
```
